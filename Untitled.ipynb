```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=[8,5])

xs = np.arange(0, 1.01, 0.01)

m_1 = 0.25
m_2 = 0.75

y1s = np.array([0 if x >= m_1 else (m_1 - x)**2 for x in xs])
y2s = np.array([0 if x <= m_2 else (m_2 - x)**2 for x in xs])


ax.plot(xs, y1s, alpha=0.5, linewidth=2, label="Different-Author Loss", linestyle="--")
ax.plot(xs, y2s, alpha=0.5, linewidth=2, label="Same-Author Loss", color="r")

ax.axvline(x=m_1, color="gray", linestyle=":")
ax.axvline(x=m_2, color="gray", linestyle=":")

plt.text(m_1 - 0.01, plt.ylim()[1] - 0.03, 'margin_diff >', rotation=0, horizontalalignment='right')
plt.text(m_2 + 0.01, plt.ylim()[1] - 0.03, '< margin_same', rotation=0, horizontalalignment='left')

ax.set_ylabel("Loss")
ax.set_xlabel("Similarity")

plt.legend()

ax.set_title("Modified Contrastive Loss")

plt.grid(True, which="both", ls="-", alpha=0.2)
plt.tight_layout(pad=False)
```
